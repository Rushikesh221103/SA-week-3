```python
train_features.fillna(1.0, inplace=True)
test_features.fillna(1.0, inplace=True)
```

```python
train_features = pd.get_dummies(train_features)
test_features = pd.get_dummies(test_features)
```

```python
train_features, test_features = train_features.align(test_features, join='left', axis=1, fill_value=0)
```

```python
train_data = train_features.merge(train_labels, on='respondent_id')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
numeric_cols = train_features.select_dtypes(include=['float64', 'int64']).columns
train_features[numeric_cols] = scaler.fit_transform(train_features[numeric_cols])
test_features[numeric_cols] = scaler.transform(test_features[numeric_cols])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
```

```python
X = train_features.drop(columns=['respondent_id'])
y_xyz = train_data['xyz_vaccine']
y_seasonal = train_data['seasonal_vaccine']
```

```python
X_train, X_val, y_train_xyz, y_val_xyz = train_test_split(X, y_xyz, test_size=0.2, random_state=42)
X_train, X_val, y_train_seasonal, y_val_seasonal = train_test_split(X, y_seasonal, test_size=0.2, random_state=42)
```

```python
rf_xyz = RandomForestClassifier(random_state=42)
rf_xyz.fit(X_train, y_train_xyz)
```

```python
rf_seasonal = RandomForestClassifier(random_state=42)
rf_seasonal.fit(X_train, y_train_seasonal)
```

```python
y_pred_xyz_proba = rf_xyz.predict_proba(X_val)[:, 1]
y_pred_seasonal_proba = rf_seasonal.predict_proba(X_val)[:, 1]
```

```python
roc_auc_xyz = roc_auc_score(y_val_xyz, y_pred_xyz_proba)
roc_auc_seasonal = roc_auc_score(y_val_seasonal, y_pred_seasonal_proba)
```

```python
print(f'ROC AUC for xyz_vaccine: {roc_auc_xyz:.4f}')
print(f'ROC AUC for seasonal_vaccine: {roc_auc_seasonal:.4f}')
print(f'Mean ROC AUC: {(roc_auc_xyz + roc_auc_seasonal) / 2:.4f}')
```

```python
# Predict probabilities on the test set
test_features = test_features.drop(columns=['respondent_id'])
test_pred_xyz_proba = rf_xyz.predict_proba(test_features)[:, 1]
test_pred_seasonal_proba = rf_seasonal.predict_proba(test_features)[:, 1]

# Prepare submission
submission = pd.DataFrame({
    'respondent_id': test_features.index,
    'xyz_vaccine': test_pred_xyz_proba,
    'seasonal_vaccine': test_pred_seasonal_proba
})

submission.to_csv('submission.csv', index=False)
```
